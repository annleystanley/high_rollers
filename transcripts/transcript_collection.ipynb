# Pulling Transcripts
- Fan-made transcripts are available on the [wiki](https://dimension20.fandom.com/wiki/Episode_Transcripts#Fantasy_High)
- I'm hoping to automate the process of downloading them so that this step doesn't have to be done manually

```python
import html5lib
import lxml
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
```

```python
res = requests.get("https://dimension20.fandom.com/wiki/Episode_Transcripts")
soup = bs(res.text, "html.parser")
transcript_links = {}
for link in soup.find_all("a"):
    url = link.get("href", "")
    if "docs.google.com" in url:
        transcript_links[link.text.strip()] = url

print(transcript_links)
```

```python
# our dataframe of urls was quite lengthy, so it was split up by season
acoc_urls = pd.read_csv('acoc_urls.csv')
aso_urls = pd.read_csv('aso_urls.csv')
fhfy_urls = pd.read_csv('fhfy_urls.csv')
fhsy_urls = pd.read_csv('fhsy_urls.csv')
tuc_urls = pd.read_csv('tuc_urls.csv')
tuc2_urls = pd.read_csv('tuc2_urls.csv')
```

```python
# downloading & cleaning html from all of our seasons
```

```python
# acoc

# list of URLs to fetch
urls = acoc_urls['URL']

# create an empty string to hold the concatenated text
text = ""

# loop to iterate over list of URLs
for url in urls:
    # fetching html
    response = requests.get(url)
    # create an instance of html2text
    h = html2text.HTML2Text()
    # converting html to text
    page_text = h.handle(response.content.decode())
    # appending text to string
    text += page_text

# write the concatenated text to a file
with open('acoc_transcripts.txt', 'w', encoding='utf-8') as f:
    f.write(text)
```

```python
acoc_text = open('acoc_transcripts.txt', encoding='utf-8').read()
aso_text = open('aso_transcripts.txt', encoding='utf-8').read()
fhfy_text = open('fhfy_transcripts.txt', encoding='utf-8').read()
fhsy_text = open('fhsy_transcripts.txt', encoding='utf-8').read()
tuc_text = open('tuc_transcripts.txt', encoding='utf-8').read()
tuc2_text = open('tuc2_transcripts.txt', encoding='utf-8').read()
```

```python
# splitting text based on page breaks, which are denoted by double newlines & then removing erroneous newlines
acoc_text_cleaned = acoc_text.split('\n\n')
acoc_df_raw = pd.DataFrame(acoc_text_cleaned)
acoc_df = pd.DataFrame(acoc_df_raw[0].str.replace('\n', ' '))
```

```python
"""function that splits text based on page breaks, which are denoted by double newlines & then removes erroneous newlines.
    then returns a dataframe containing every line from the script"""
def script_cleaning(season):
    season_text_cleaned = season.split('\n\n')
    season_df_raw = pd.DataFrame(season_text_cleaned)
    season_df = pd.DataFrame(season_df_raw[0].str.replace('\n', ' '))
    return season_df
```

```python
# converting scripts into dataframes
acoc_df = script_cleaning(acoc_text)
aso_df = script_cleaning(aso_text)
fhfy_df = script_cleaning(fhfy_text)
fhsy_df = script_cleaning(fhsy_text)
tuc_df = script_cleaning(tuc_text)
tuc2_df = script_cleaning(tuc2_text)
```

```python
# exporting scripts as csvs
acoc_df.to_csv('acoc_df.csv', index=False)
aso_df.to_csv('aso_df.csv', index=False)
fhfy_df.to_csv('fhfy_df.csv', index=False)
fhsy_df.to_csv('fhsy_df.csv', index=False)
tuc_df.to_csv('tuc_df.csv', index=False)
tuc2_df.to_csv('tuc2_df.csv', index=False)
```
